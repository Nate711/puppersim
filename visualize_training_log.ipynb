```python
import pandas as pd
import numpy as np
import bokeh.plotting
import bokeh.models
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
```

```python
df = pd.read_csv('data/log.txt', sep='\t')
df
output_notebook()
fig = bokeh.plotting.figure(width=500,height=300)
cds = ColumnDataSource(df)
fig.line(source=cds, x='Iteration', y='AverageReward')
fig.yaxis.axis_label = 'AverageReward'
fig.xaxis.axis_label = 'Iteration'
bokeh.plotting.show(fig)
```
